### Importing necessary libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
```

## Loading the dataset

```python
from tensorflow.keras.datasets import mnist
```

```python
randomstate=10
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

### Visualising few samples

```python
plt.figure(figsize=(9,9))
for index,(image,label) in enumerate(zip(x_train[:9],y_train[:9])):
    plt.subplot(3,3,index+1).imshow(image,interpolation ='bicubic')
    plt.title(f'This figure shows {label}')
    plt.tight_layout()
```

### making logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import metrics
```

```python
logreg=LogisticRegression()
```

```python
x_train=x_train.reshape(60000,-1)
x_test=x_test.reshape(10000,-1)
```

```python
logreg.fit(x_train,y_train)
```

```python
score=logreg.score(x_test,y_test)
score
```

```python
y_pred=logreg.predict(x_test)
```

```python
cm=metrics.confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True,linewidths=0.5,square=True,xticklabels=True,yticklabels=True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'The overall accuracy is {score}')
```

```python
print(metrics.classification_report(y_test,y_pred))
```
